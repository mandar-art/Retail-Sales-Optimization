#### Here we can see apart from 1 null value in ship mode, there are other two as Not Available and Unknown which can be considered as Null. So we will use pandas to treat these values as null only.

```python
df = pd.read_csv('Downloads/orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df.isnull().sum()
```

```python
df['Ship Mode'].unique()
```

#### So now we will fix the column names for handling dataset in better way

```python
#Rename columns

df.columns
```

```python
df.columns = df.columns.str.lower()
#df.columns                                        #All column names changed to lower case
```

```python
df.columns = df.columns.str.replace(' ' , '_')
#df.columns                                       #All spaces changed with underscore
```

#### Now we have to create new columns for better insights on sales and discount

```python
#Derive new columns sales, discount, profit

df['discount'] = df['list_price']*df['discount_percent']/100
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

#### Now we can drop few columns as cost price, list price discount percent

```python
df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace = True)    

#We use inplace=True because without it columns will not be dropped in actual dataset
```

#### Now lets check for datatypes

```python
df.dtypes
```

```python
#We have to change datatype for order_date

df['order_date'] = pd.to_datetime(df['order_date'])
```
